TASK 2:  Tic tac toe game using python

```python
import random

# Initialize the board
board = [" " for _ in range(9)]

# Function to print the board
def print_board():
    for row in [board[i:i+3] for i in range(0, 9, 3)]:
        print(" | ".join(row))
        print("-" * 9)

# Function to check for a win
def check_win(player):
    # Check rows, columns, and diagonals
    win_patterns = [(0, 1, 2), (3, 4, 5), (6, 7, 8), (0, 3, 6), (1, 4, 7), (2, 5, 8), (0, 4, 8), (2, 4, 6)]
    for pattern in win_patterns:
        if board[pattern[0]] == board[pattern[1]] == board[pattern[2]] == player:
            return True
    return False

# Function to check for a tie
def check_tie():
    return " " not in board

# Function for bot's move (random)
def bot_move():
    available_moves = [i for i in range(9) if board[i] == " "]
    return random.choice(available_moves)

# Function for player vs. bot game
def play_vs_bot():
    player_symbol = "X"
    bot_symbol = "O"

    while True:
        print_board()

        # Player's turn
        print(f"Player's turn. Enter a position (1-9): ")
        move = input()

        if not move.isdigit() or int(move) < 1 or int(move) > 9 or board[int(move) - 1] != " ":
            print("Invalid move. Try again.")
            continue

        board[int(move) - 1] = player_symbol

        if check_win(player_symbol):
            print_board()
            print("Player wins! Congratulations!")
            break
        elif check_tie():
            print_board()
            print("It's a tie!")
            break

        # Bot's turn
        bot_position = bot_move()
        board[bot_position] = bot_symbol

        if check_win(bot_symbol):
            print_board()
            print("Bot wins! Better luck next time!")
            break
        elif check_tie():
            print_board()
            print("It's a tie!")
            break

# Function for player vs. player game
def play_vs_player():
    player1_symbol = "X"
    player2_symbol = "O"
    current_player = player1_symbol

    while True:
        print_board()

        print(f"Player {current_player}'s turn. Enter a position (1-9): ")
        move = input()

        if not move.isdigit() or int(move) < 1 or int(move) > 9 or board[int(move) - 1] != " ":
            print("Invalid move. Try again.")
            continue

        board[int(move) - 1] = current_player

        if check_win(current_player):
            print_board()
            print(f"Player {current_player} wins! Congratulations!")
            break
        elif check_tie():
            print_board()
            print("It's a tie!")
            break

        # Switch players
        current_player = player2_symbol if current_player == player1_symbol else player1_symbol

# Main menu
while True:
    print("Welcome to Tic Tac Toe!")
    print("1. Play against Bot")
    print("2. Play between two Players")
    print("3. Quit")
    choice = input("Enter your choice (1/2/3): ")

    if choice == "1":
        board = [" " for _ in range(9)]  # Reset the board
        play_vs_bot()
    elif choice == "2":
        board = [" " for _ in range(9)]  # Reset the board
        play_vs_player()
    elif choice == "3":
        break
    else:
        print("Invalid choice. Please select 1, 2, or 3.")
```
